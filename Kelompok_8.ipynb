```python
!pip install pdfplumber nltk

import pdfplumber
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import nltk
from collections import defaultdict

# Unduh resource yang dibutuhkan nltk
nltk.download('punkt')
nltk.download('stopwords')

# Inisialisasi stemmer dan stopwords
stemmer = PorterStemmer()
stop_words = set(stopwords.words('indonesian'))  # Stopwords dalam bahasa Indonesian

file_path = "/content/akutansi.pdf"  # Nama file
inverted_index = defaultdict(list)  # Struktur data untuk inverted index

# Baca file PDF dan proses teks
with pdfplumber.open(file_path) as pdf:
    for page_num, page in enumerate(pdf.pages):
        text = page.extract_text()
        if text:
            # Mengubah teks menjadi huruf kecil
            text = text.lower()
            # Menghilangkan tanda baca dan angka
            text = re.sub(r'[^\w\s]', '', text)  # Menghilangkan tanda baca
            text = re.sub(r'\d+', '', text)      # Menghilangkan angka
            # Tokenisasi teks
            tokens = word_tokenize(text)
            # Filtering dan Stemming
            filtered_tokens = [stemmer.stem(token) for token in tokens if token not in stop_words]
            # Membuat inverted index
            for token in filtered_tokens:
                if page_num not in inverted_index[token]:  # Cegah duplikasi
                    inverted_index[token].append(page_num)

# Tampilkan hasil inverted index
for token, pages in inverted_index.items():
    print(f"{token}: {pages}")
```
